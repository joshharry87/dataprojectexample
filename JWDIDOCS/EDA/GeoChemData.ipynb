````python

ppmcols = []
for col in data.columns:
    if "ppm" in col:
        ppmcols.append(col)



def sum(row):
    total = 0
    print(row['Ag_ppm'])
    for col in ppmcols:
        if "ppm" in col:
            total += row[col]
            
    return total

data["ppm_sum"] = data.apply(lambda row: sum(row), axis=1)

````

````python


for col in ppmcols:
    data[f"{col}_%"] = data.apply(lambda x: (x[col]/x['ppm_sum'])*100 if x['ppm_sum'] > 0  else 0, axis=1)
````

````python

def find_best_element(row):
    base_index = 0
    base = 0
    index = 0
    element = ""
    for col in ppmcols:
        if row[f"{col}_%"] > base:
            base = row[f"{col}_%"]
            base_index = index
            
        index +=1
    return ppmcols[base_index].split("_")[0]



def all_elements_above_10(row):
    elements = ""
    for col in ppmcols:
        if row[f"{col}_%"] > 10:
            elements += col.split("_")[0] + ","
    
    return elements[0:-1]

def all_present_elements(row):
    elements = ""
    for col in ppmcols:
        if row[f"{col}_%"] > 0:
            elements += col.split("_")[0] + ","
    
    return elements[0:-1]

for col in ppmcols:
    data[f"Top_1_Elements"] = data.apply(lambda row: find_best_element(row), axis=1)
    data[f"Elements_over_10"] = data.apply(lambda row: all_elements_above_10(row), axis=1)
    data[f"all_present_elements"] = data.apply(lambda row: all_present_elements(row), axis=1)
    
````
